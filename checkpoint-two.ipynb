```python
# concatenate the datasets
# overview of datasets composition in terms of food groups
# which food groups are most and least represented
df_combined = pd.concat([df_test, df_train], ignore_index = True)
#set up the plotting style
plt.style.use('ggplot')
# Distribution of Food Groups
plt.figure(figsize = (12,6))
df_combined['FoodGroup'].value_counts().plot(kind ='bar')
plt.title('Disribution of Food Groups')
plt.xlabel('Food Group')
plt.ylabel('Count')
plt.xticks(rotation = 45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Distribution of Energy content
# Looking for peaks in the distribution,energy levels in food
# understanding the overall nutritional profile of the foods.
plt.figure(figsize = (10,6))
sns.histplot(df_combined['Energy_kcal'], kde = True)
plt.title('Distribution of Energy Content')
plt.xlabel('Energy (Kcal)')
plt.show()
          
```

```python
# Define the nutrients to include
# overview of how different nutrients are distributed across all food items.
# identify ranges, medians, outliers for each nutrient.
#looking for unusual patterns or outliers.
nutrients = ['Protein_g', 'Fat_g', 'Carb_g', 'VitA_mcg', 'VitB6_mg','VitB12_mcg', 'VitE_mg']
# Create the boxplot
plt.figure (figsize =(12,10))
df_combined[nutrients].boxplot()
# customize the plot
plt.title('Nutrient Composition', fontsize = 16)
plt.ylabel('Amount', fontsize = 12)
plt.xlabel('Nutrients', fontsize = 12)
plt.xticks(rotation = 45, ha ='right')
plt.yscale('log') # using log scale due to large differences in ranges
#Add a grid for better readability
plt.grid(True, axis ='y', linestyle = '--', alpha = 0.7)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of Protien Vs Fat content
# looking for relationship between protien and fat content across different food groups
plt.figure(figsize =(12,8))
sns.scatterplot(x = 'Protein_g', y = 'Fat_g', hue = 'FoodGroup', data = df_combined, alpha = 0.6)
plt.title('Protein vs Fat Content by Food Group')
plt.xlabel('Protein (g)')
plt.ylabel('Fat (g)')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')
plt.tight_layout()
plt.show()

           
```

```python
# Average Vitamin Content (% USRDA)
# looking for which vitamins are most prevalent in the foods
plt.figure(figsize =(12,6))
vitamin_cols = [col for col in df_combined.columns if col.startswith('Vit') and col.endswith('_USRDA')] 
df_combined[vitamin_cols].mean().sort_values(ascending = False).plot(kind = 'bar')
plt.title('Average Vitamin Content (% USRDA)')
plt.ylabel('Percentage of USRDA')
plt.xticks(rotation = 45, ha ='right')
plt.tight_layout()
plt.show()
```

```python
# Select a few food groups
selected_groups = ['Beef Products', 'Pork Products', 'Finfish and Shellfish Products', 'Vegetables and Vegetable Products']
# Create a scatter plot
plt.figure(figsize = (12,8))
for group in selected_groups:
    group_data = df_combined[df_combined['FoodGroup'] == group]
    plt.scatter(group_data['Protein_g'],group_data['Fat_g'], label = group, alpha = 0.7)
plt.xlabel('Protein (g)')
plt.ylabel('Fat (g)')
plt.title('Protein vs Fat Content in selected Food Groups')
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.7)
# Add a line for  1:1 protein to fat ratio
max_value = max(df_combined['Protein_g'].max(),df_combined['Fat_g'].max())
plt.plot([0, max_value], [0, max_value], 'r--', label = '1:1 Protein:Fat ratio')
plt.legend()
plt.tight_layout()
plt.show()
                
          
    
                   
```

```python
# Define the daily recommended value for protein (in grams)
# Calculate the median protein content for each food group
# Create a custom order for the food groups based on median protein content

daily_protein_value = 50 
median_protein = df_combined.groupby('FoodGroup')['Protein_g'].median().sort_values(ascending=False)
custom_order = median_protein.index

# Create the boxplot with the custom order
# Add a horizontal line for the daily nutritional value
# Add a text label for the reference line

plt.figure(figsize=(12, 6))
ax = sns.boxplot(x='FoodGroup', y='Protein_g', data=df_combined, order=custom_order)
ax.axhline(y=daily_protein_value, color='r', linestyle='--', linewidth=2)
ax.text(ax.get_xlim()[1], daily_protein_value, 'Daily Value', 
        horizontalalignment='left', verticalalignment='bottom', 
        color='r', fontweight='bold')

plt.xticks(rotation=90)
plt.title('Protein Content by Food Group (Sorted by Median Protein Content)')
plt.xlabel('Food Group')
plt.ylabel('Protein (g)')

# Adjust the y-axis to ensure the reference line is visible
y_max = max(df_combined['Protein_g'].max(), daily_protein_value) * 1.1
plt.ylim(0, y_max)

plt.tight_layout()
plt.show()
```

```python
# Define the daily recommended value for carbohydrates (in grams)
# Based on a 2000 calorie diet, adjust if needed
# Use the correct column name for carbohydrates
# Calculate the median carbohydrate content for each food group
# Create a custom order for the food groups based on median carbohydrate content
# Identify food groups with median carbohydrate content > 50g (adjust this threshold as needed)


daily_carb_value = 275 
carb_column = 'Carb_g'
median_carb = df_combined.groupby('FoodGroup')[carb_column].median().sort_values(ascending=False)
custom_order = median_carb.index
high_carb_groups = median_carb[median_carb > 50].index

# Create a color palette: orange for high carb, gray for others
# Create the boxplot with the custom order and color palette
# Add a horizontal line for the daily nutritional value
# Add a text label for the reference line

color_palette = {group: '#FFA500' if group in high_carb_groups else '#A9A9A9' for group in custom_order}
plt.figure(figsize=(14, 6))
ax = sns.boxplot(x='FoodGroup', y=carb_column, data=df_combined, order=custom_order, 
                 hue='FoodGroup', palette=color_palette, legend=False)
ax.axhline(y=daily_carb_value, color='r', linestyle='--', linewidth=2)

ax.text(ax.get_xlim()[1], daily_carb_value, 'Daily Value', 
        horizontalalignment='left', verticalalignment='bottom', 
        color='r', fontweight='bold')

plt.xticks(rotation=90)
plt.title('Carbohydrate Content by Food Group (Sorted by Median Carbohydrate Content)')
plt.xlabel('Food Group')
plt.ylabel('Carbohydrate (g)')

# Adjust the y-axis to ensure the reference line is visible
y_max = max(df_combined[carb_column].max(), daily_carb_value) * 1.1
plt.ylim(0, y_max)

# Create a custom legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#FFA500', edgecolor='#FFA500', label='Median Carb > 50g'),
                   Patch(facecolor='#A9A9A9', edgecolor='#A9A9A9', label='Median Carb  50g')]
plt.legend(handles=legend_elements, title='Food Groups', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Define the daily recommended value for fats (in grams)
# Based on a 2000 calorie diet, adjust if needed
# Calculate the median fat content for each food group
# Create a custom order for the food groups based on median fat content
# Identify food groups with median fat content > 20g 

daily_fat_value = 78
fat_column = 'Fat_g'
median_fat = df_combined.groupby('FoodGroup')[fat_column].median().sort_values(ascending=False)
custom_order = median_fat.index
high_fat_groups = median_fat[median_fat > 20].index

# Create a color palette: red for high fat, gray for others
# Create the boxplot with the custom order and color palette
# Add a horizontal line for the daily nutritional value
# Add a text label for the reference line

color_palette = {group: '#FF4500' if group in high_fat_groups else '#A9A9A9' for group in custom_order}
plt.figure(figsize=(14, 6))
ax = sns.boxplot(x='FoodGroup', y=fat_column, data=df_combined, order=custom_order, 
                 hue='FoodGroup', palette=color_palette, legend=False)
ax.axhline(y=daily_fat_value, color='r', linestyle='--', linewidth=2)
ax.text(ax.get_xlim()[1], daily_fat_value, 'Daily Value', 
        horizontalalignment='left', verticalalignment='bottom', 
        color='r', fontweight='bold')

plt.xticks(rotation=90)
plt.title('Fat Content by Food Group (Sorted by Median Fat Content)')
plt.xlabel('Food Group')
plt.ylabel('Fat (g)')

# Adjust the y-axis to ensure the reference line is visible
y_max = max(df_combined[fat_column].max(), daily_fat_value) * 1.1
plt.ylim(0, y_max)

# Create a custom legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#FF4500', edgecolor='#FF4500', label='Median Fat > 20g'),
                   Patch(facecolor='#A9A9A9', edgecolor='#A9A9A9', label='Median Fat  20g')]
plt.legend(handles=legend_elements, title='Food Groups', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Define the daily recommended value for fiber (in grams)
# Based on a general recommendation for adults, adjust if needed
# Calculate the median fiber content for each food group
# Create a custom order for the food groups based on median fiber content
# Identify food groups with median fiber content > 5g 


daily_fiber_value = 25 
fiber_column = 'Fiber_g'
median_fiber = df_combined.groupby('FoodGroup')[fiber_column].median().sort_values(ascending=False)
custom_order = median_fiber.index
high_fiber_groups = median_fiber[median_fiber > 5].index

# Create a color palette: green for high fiber, gray for others
# Create the boxplot with the custom order and color palette
# Add a horizontal line for the daily nutritional value
# Add a text label for the reference line

color_palette = {group: '#32CD32' if group in high_fiber_groups else '#A9A9A9' for group in custom_order}
plt.figure(figsize=(14, 6))
ax = sns.boxplot(x='FoodGroup', y=fiber_column, data=df_combined, order=custom_order, 
                 hue='FoodGroup', palette=color_palette, legend=False)
ax.axhline(y=daily_fiber_value, color='r', linestyle='--', linewidth=2)
ax.text(ax.get_xlim()[1], daily_fiber_value, 'Daily Value', 
        horizontalalignment='left', verticalalignment='bottom', 
        color='r', fontweight='bold')

plt.xticks(rotation=90)
plt.title('Fiber Content by Food Group (Sorted by Median Fiber Content)')
plt.xlabel('Food Group')
plt.ylabel('Fiber (g)')

# Adjust the y-axis to ensure the reference line is visible
y_max = max(df_combined[fiber_column].max(), daily_fiber_value) * 1.1
plt.ylim(0, y_max)

# Create a custom legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#32CD32', edgecolor='#32CD32', label='Median Fiber > 5g'),
                   Patch(facecolor='#A9A9A9', edgecolor='#A9A9A9', label='Median Fiber  5g')]
plt.legend(handles=legend_elements, title='Food Groups', loc='upper right')

plt.tight_layout()
plt.show()
```
